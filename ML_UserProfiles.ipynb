```python
import pandas

ratings_features = ["userId", "movieId", "rating", "timestamp"]
ratings_dataframe = pandas.read_csv("u.data", sep = "\t", names = ratings_features,
                                    encoding = "ISO-8859-1")
```

```python
movie_features = ["movieId", "movieTitle", "Release", "videoRelease", "IMDB URL", "unknown",
                  "Action", "Adventure", "Animation", "Childrens", "Comedy", "Crime",
                  "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical",
                  "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]

movies_dataframe = pandas.read_csv("u.item", sep = "|", names = movie_features,
                                    encoding = "ISO-8859-1")

movies_dataframe
```

```python
user_features = ["userId", "age", "gender", "occupation", "zipCode"]
users_dataframe = pandas.read_csv("u.user", sep = "|", names = user_features,
                                    encoding = "ISO-8859-1")

users_dataframe
```

```python
NUMBER_OF_GROUPS = 5

users_dataframe["age group"] = pandas.qcut(users_dataframe["age"],
                                           q = NUMBER_OF_GROUPS,
                                           precision = 0)

users_dataframe
```

```python
merged_dataframes = pandas.merge(pandas.merge(ratings_dataframe,
                                              users_dataframe[["userId",
                                                               "age group",
                                                               "gender",
                                                               "occupation"]],
                                              on = "userId",
                                              how = "left"),
                                 movies_dataframe,
                                 on = "movieId",
                                 how = "left")
merged_dataframes
```

```python
# Drop columns that are not useful for ML model. Extra info thats not needed can / will affect results
merged_dataframes.drop(["movieId", "movieTitle", "Release", "timestamp", "unknown", "IMDB URL", "videoRelease"], axis = 1,
                       inplace = True)

merged_dataframes
```

```python
merged_dataframes ["age group"] = pandas.Categorical(merged_dataframes["age group"])

merged_dataframes
```

```python
age_group_dummies = pandas.get_dummies(merged_dataframes["age group"])

age_group_dummies
```

```python
merged_dataframes["gender"] = pandas.Categorical(merged_dataframes["gender"])

gender_dummies = pandas.get_dummies(merged_dataframes["gender"])

gender_dummies
```

```python
merged_dataframes["occupation"] = pandas.Categorical(merged_dataframes["occupation"])

occupation_dummies = pandas.get_dummies(merged_dataframes["occupation"])

occupation_dummies
```

```python
merged_dataframes = pandas.concat([merged_dataframes,
                                   age_group_dummies,
                                   gender_dummies,
                                   occupation_dummies],
                                  axis = 1)

merged_dataframes
```

```python
merged_dataframes.drop(["age group", "gender", "occupation"], axis = 1,
                       inplace = True)

merged_dataframes
```

```python
# Check to see if there are any null values as they're not required
merged_dataframes.isnull().sum()
```

```python
from sklearn.linear_model import Ridge

from sklearn.model_selection import GridSearchCV

#Create base ridge model
ridge_model = Ridge()

alpha = []

ALPHA_LENGTH = 7

import numpy

# Create list of estimator numbers
for i in range(ALPHA_LENGTH):

  alpha.extend(numpy.arange(10**(i-5), 10**(i-4), 10**(i-5)*2))

alpha
```

```python
# Create dictionary of estimators
parameters = { "alpha": alpha }

NUMBER_OF_FOLDS = 5

# Used to evaluate how well a model perfoms and compare different models using the estimator
ridge_cross_validation = GridSearchCV(estimator = ridge_model,
                                      param_grid= parameters,
                                      scoring = "neg_mean_absolute_error",
                                      cv = NUMBER_OF_FOLDS,
                                      return_train_score=True,
                                      verbose = 1)

COLUMN_AXIS = 1

# Independent data
X = merged_dataframes.drop("rating",
                           axis = COLUMN_AXIS)

# Dependent data
y = merged_dataframes.rating

from sklearn.model_selection import train_test_split

# Split in to groups with a test size of 20% (i.e./ 20k)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

X_train
```

```python
X_train.columns = X_train.columns.astype(str)

# function used to train (.fit == training)
ridge_cross_validation.fit(X_train, y_train)
```

```python
# Find best value that matchs our model / data
ridge_cross_validation.best_estimator_
```

```python
best_alpha = 1e-05

# Create most accurate ridge  using the best estimator
best_alpha_ridge = Ridge(alpha = best_alpha)

# Train data again using ridge to get estimator back out. Confirming corrrect
best_alpha_ridge.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

X_test.columns = X_test.columns.astype(str)

# Find error of model using the NEW data from test. (i.e/ not seen before)
error = mean_squared_error(y_test, best_alpha_ridge.predict(X_test))

error
```

```python
ridge_results_dataframe = pandas.DataFrame({
    "Features": X_train.columns,
    "Coefficient": best_alpha_ridge.coef_
})

ridge_results_dataframe
```

```python
import matplotlib.pyplot as pyplot

figure, axes = pyplot.subplots(figsize = [7,15])

import seaborn

seaborn.barplot(x = "Coefficient",
                y= "Features",
                ax = axes,
                data = ridge_results_dataframe)
```
