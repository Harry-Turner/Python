```python
import pandas

ratings_features = ["user_id", "movieId", "rating", "timestamp"]

ratings_dataframe = pandas.read_csv("ratings.csv",
                                    names = ratings_features)

ratings_dataframe
```

```python
FIRST_INDEX_ROW = 0
ratings_dataframe = ratings_dataframe.drop(ratings_dataframe.index[FIRST_INDEX_ROW])

ratings_dataframe

ratings_dataframe.info()
```

```python
ratings_dataframe = ratings_dataframe.astype("float")

ratings_dataframe.info()
```

```python
movies_dataframe = pandas.read_csv("movies.csv")
movies_dataframe.info()
```

```python
movie_titles_dataframe = movies_dataframe[["movieId", "title"]]
movie_titles_dataframe
```

```python
movie_titles_dataframe["movieId"] = movie_titles_dataframe["movieId"].astype(str).astype(float)
movie_titles_dataframe.info()
```

```python
# Combine the 2 dataframes
merged_dataframe = pandas.merge(ratings_dataframe, movie_titles_dataframe, on="movieId")
merged_dataframe
```

```python
# Show how many separate ratings a movie got
merged_dataframe.groupby("movieId")["rating"].count().sort_values(ascending = False)
```

```python
# Create table to show result "crosstabulation", in this case the titel against user_id and rating score as the value
crosstab = merged_dataframe.pivot_table(values = "rating",
                            index = "user_id",
                            columns = "title",
                            fill_value = 0)

crosstab
```

```python
# Transpose the table (Swap the axis)
x = crosstab.T
x
```

```python
# Step needed before a correlation matrix. Does stuff with values above and puts in to array
from sklearn.decomposition import TruncatedSVD

NUMBER_OF_COMPONENTS = 12

singular_value_decomposition = TruncatedSVD(n_components = NUMBER_OF_COMPONENTS,
             random_state = 1)

matrix = singular_value_decomposition.fit_transform(x)
matrix
```

```python
# Convert table above in to correlation coefficients - correlation between one movie and another
import numpy

correlation_matrix = numpy.corrcoef(matrix)
correlation_matrix
```

```python
movie_titles = crosstab.columns # Get all movie titles
movies_list = list(movie_titles) # Display movie titles in list

example_movie_index = movies_list.index("Batman: Year One (2011)") # Choose movie to recommend for
example_correlations = correlation_matrix[example_movie_index] # Get all the coefficients for this movie

# Show all movie titles which are correlated between MIN & MAX for chosen film
MAXIMUM_CORRELATION = 1.0
MINIMUM_CORRELATION = 0.9
list(movie_titles[(example_correlations < MAXIMUM_CORRELATION) & (example_correlations > MINIMUM_CORRELATION)])
```
